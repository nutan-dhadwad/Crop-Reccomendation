# Lasso & Ridge Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from sklearn.linear_model import Lasso, Ridge

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error,  root_mean_squared_error
```

```python
x = np.random.randint(0,100,300)
y = -3 * x + np.random.rand(300) * 150
```

```python
df=pd.DataFrame({"x":x, "y":y})
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[["x"]], df["y"], 
                                                    train_size=0.75, 
                                                    random_state=42)
```

```python
plt.scatter(x_train, y_train, label="train")
plt.scatter(x_test, y_test, label="test")
plt.legend()
plt.show()
```

```python

Lasso_Model= Lasso(alpha= 0.5)
Ridge_Model=Ridge(alpha= 2)
```

```python
Lasso_Model.fit(x_train, y_train)
Ridge_Model.fit(x_train, y_train)
```

```python
y_pred_l= Lasso_Model.predict(x_test)
y_pred_r= Ridge_Model.predict(x_test)
```

```python
print("lasso model:training : R2", r2_score(y_true=y_train, y_pred=Lasso_Model.predict(x_train)))
print("ridge: testing : R2", r2_score(y_true=y_test, y_pred=Lasso_Model.predict(x_test)))


print("ridge: trainig :r2",r2_score(y_true=y_train, y_pred=Ridge_Model.predict(x_train)))
print("ridge: testing :r2",r2_score(y_true= y_test, y_pred=Ridge_Model.predict(x_test)))
```

```python
x_line= np.linspace(0,100, 1000).reshape(-1,1)
y_line_l= Lasso_Model.predict(x_line)
y_line_r=Ridge_Model.predict(x_line)


plt.scatter(x_train, y_train, label="training")
plt.scatter(x_test, y_test, label="testing")
plt.plot(x_line, y_line_l, label="Lasso", alpha=0.5)
plt.plot(x_line, y_line_r, label="Ridge", alpha=0.5)
plt.legend()
plt.show()
```
