```python
import numpy as np

def removeoutliers(df, col):
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    IQR = q3 - q1
    LW = q1 - 1.5 * IQR
    UW = q3 + 1.5 * IQR
    return df[(df[col] >= LW) & (df[col] <= UW)]

new_df = removeoutliers(df, "selling_price")

```

```python
plt.boxplot(new_df["km_ns_sqrt"])
plt.show()
```
