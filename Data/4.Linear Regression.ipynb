```python
import numpy as np 
import pandas as pd 
```

```python
x = np.random.randint(0,100,100)
y = 3 * x + (np.random.rand(100) * 100)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df=pd.DataFrame({
    "x":x,
    "y":y})
df
```

```python
x= df[["x"]]
y= df[["y"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2)
```

```python
plt.scatter(x_train, y_train, label="Train")
plt.scatter(x_test, y_test, label="Test")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_, lr.intercept_
```

```python
# y= 2.84 * x + 54.56
```

```python
y_pred= lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_test, y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(r2_score(y_test, y_pred))
```

```python
print(mean_absolute_error(y_train, lr.predict(x_train)))
print(root_mean_squared_error(y_train, lr.predict(x_train)))
print(r2_score(y_train, lr.predict(x_train)))
```

```python
x_line=np.linspace(0,100,1000).reshape(-1,1)
y_line=lr.predict(x_line)
```

```python
plt.scatter(x_train, y_train, label="train")
plt.scatter(x_test, y_test, label="test")
plt.plot(x_line, y_line, label="prediction line")
plt.legend()
plt.show()
```
