```python
import pandas as pd
import seaborn as sns


from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
df= pd.DataFrame(data=wine["data"], columns=wine["feature_names"])
df["class"]= wine["target"]


df.sample(10)
```

```python
X = df.drop("class", axis=1)
y= df["class"]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size= 0.8, random_state= 42, stratify= y)
```

```python
y_train
```

```python
y_test
```

```python
mns= MinMaxScaler()
X_train_scaled =mns.fit_transform(X_train)
X_test_scaled= mns.transform(X_test)
```

```python
X_scaled_train_df= pd.DataFrame(X_train_scaled, columns=wine["feature_names"])
```

```python
X_scaled_test_df= pd.DataFrame(X_test_scaled, columns=wine["feature_names"])
```

```python
pca= PCA(n_components=7)
X_train_pca= pca.fit_transform(X_scaled_train_df)
X_test_pca= pca.transform(X_scaled_test_df)
```

```python
X_test_pca
```

```python
X_train_pca
```

```python
dtc=DecisionTreeClassifier()
dtc_scaled= DecisionTreeClassifier()
dtc_pca= DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
dtc_scaled.fit(X_scaled_train_df, y_train)
dtc_pca.fit(X_train_pca, y_train)
```

```python
print("accuracy on driginal data:")
print("train :", accuracy_score(y_true= y_train, y_pred=dtc.predict(X_train)))
print("test :", accuracy_score(y_true= y_test, y_pred=dtc.predict(X_test)))


print("accuracy on scaled data:")
print("train :", accuracy_score(y_true= y_train, y_pred=dtc_scaled.predict(X_scaled_train_df)))
print("train :", accuracy_score(y_true= y_test, y_pred=dtc_scaled.predict(X_scaled_test_df)))


print("accuracy on PCA data:")
print("train :", accuracy_score(y_true= y_train, y_pred=dtc_pca.predict(X_train_pca)))
print("train :", accuracy_score(y_true= y_test, y_pred=dtc_pca.predict(X_test_pca)))

```
