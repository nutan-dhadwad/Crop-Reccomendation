# Preprocessing

# Encoding

```python
for col in df.select_dtypes(include="object").columns:
    print(col,":",df[col].unique())
```

```python
le_geo=LabelEncoder()

le_gender=LabelEncoder()

oe_crdtype= OrdinalEncoder(categories=[["SILVER","GOLD", "DIAMOND", "PLATINUM"]])
 

#geography col
xtrain["Geography"] = le_geo.fit_transform(xtrain["Geography"])
xval["Geography"] = le_geo.transform(xval["Geography"])
xtest["Geography"] = le_geo.transform(xtest["Geography"])


#gender
xtrain["Gender"] = le_geo.fit_transform(xtrain["Gender"])
xval["Gender"] = le_geo.transform(xval["Gender"])
xtest["Gender"] = le_geo.transform(xtest["Gender"])

#card type
xtrain["Card Type"] = oe_crdtype.fit_transform(xtrain[["Card Type"]])
xval["Card Type"] = oe_crdtype.transform(xval[["Card Type"]])
xtest["Card Type"] = oe_crdtype.transform(xtest[["Card Type"]])
```

```python
# Age, CreditScore, Point Earned, Balance, EstimatedSalary
mms = MinMaxScaler()
xtrain[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]] \
            = mms.fit_transform(xtrain[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]])
xval[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]] \
            = mms.transform(xval[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]])
xtest[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]] \
            = mms.transform(xtest[["Age", "CreditScore", "Point Earned", "Balance", "EstimatedSalary"]])
```

# Training Model

```python
lr = LogisticRegression()
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
abc = AdaBoostClassifier()
gbc = GradientBoostingClassifier()
xgb = XGBClassifier()
gnb = GaussianNB()
svc = SVC()
```

```python
lr.fit(xtrain, ytrain)
print(f"#Train :{lr.score(xtrain, ytrain)} #Val :{lr.score(xval, yval)} #Test :{lr.score(xtest,ytest)} ")

dtc.fit(xtrain, ytrain)
print(f"#Train :{dtc.score(xtrain, ytrain)} #Val :{dtc.score(xval, yval)} #Test :{dtc.score(xtest,ytest)} ")

rfc.fit(xtrain, ytrain)
print(f"#Train :{rfc.score(xtrain, ytrain)} #Val :{rfc.score(xval, yval)} #Test :{rfc.score(xtest,ytest)} ")

abc.fit(xtrain, ytrain)
print(f"#Train :{abc.score(xtrain, ytrain)} #Val :{abc.score(xval, yval)} #Test :{abc.score(xtest,ytest)} ")

gbc.fit(xtrain, ytrain)
print(f"#Train :{gbc.score(xtrain, ytrain)} #Val :{gbc.score(xval, yval)} #Test :{gbc.score(xtest,ytest)} ")

xgb.fit(xtrain, ytrain)
print(f"#Train :{xgb.score(xtrain, ytrain)} #Val :{xgb.score(xval, yval)} #Test :{xgb.score(xtest,ytest)} ")

gnb.fit(xtrain, ytrain)
print(f"#Train :{gnb.score(xtrain, ytrain)} #Val :{gnb.score(xval, yval)} #Test :{gnb.score(xtest,ytest)} ")

svc.fit(xtrain, ytrain)
print(f"#Train :{svc.score(xtrain, ytrain)} #Val :{svc.score(xval, yval)} #Test :{svc.score(xtest,ytest)} ")
```

```python
lr.fit(xtrain, ytrain)
ypred_lr = lr.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_lr)} #Prec :{precision_score(ytest, ypred_lr, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_lr, average='weighted')} #F1 :{f1_score(ytest, ypred_lr, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_lr)}")

dtc.fit(xtrain, ytrain)
ypred_dtc = dtc.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_dtc)} #Prec :{precision_score(ytest, ypred_dtc, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_dtc, average='weighted')} #F1 :{f1_score(ytest, ypred_dtc, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_dtc)}")

rfc.fit(xtrain, ytrain)
ypred_rfc = rfc.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_rfc)} #Prec :{precision_score(ytest, ypred_rfc, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_rfc, average='weighted')} #F1 :{f1_score(ytest, ypred_rfc, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_rfc)}")

abc.fit(xtrain, ytrain)
ypred_abc = abc.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_abc)} #Prec :{precision_score(ytest, ypred_abc, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_abc, average='weighted')} #F1 :{f1_score(ytest, ypred_abc, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_abc)}")

gbc.fit(xtrain, ytrain)
ypred_gbc = gbc.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_gbc)} #Prec :{precision_score(ytest, ypred_gbc, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_gbc, average='weighted')} #F1 :{f1_score(ytest, ypred_gbc, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_gbc)}")

xgb.fit(xtrain, ytrain)
ypred_xgb = xgb.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_xgb)} #Prec :{precision_score(ytest, ypred_xgb, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_xgb, average='weighted')} #F1 :{f1_score(ytest, ypred_xgb, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_xgb)}")

gnb.fit(xtrain, ytrain)
ypred_gnb = gnb.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_gnb)} #Prec :{precision_score(ytest, ypred_gnb, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_gnb, average='weighted')} #F1 :{f1_score(ytest, ypred_gnb, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_gnb)}")

svc.fit(xtrain, ytrain)
ypred_svc = svc.predict(xtest)
print(f"#Acc :{accuracy_score(ytest, ypred_svc)} #Prec :{precision_score(ytest, ypred_svc, average='weighted', zero_division=0)} "
      f"#Rec :{recall_score(ytest, ypred_svc, average='weighted')} #F1 :{f1_score(ytest, ypred_svc, average='weighted')} "
      f"#CM :{confusion_matrix(ytest, ypred_svc)}")
```
